# Biopython

Write the following five functions to analyze global alignments between two sequences using Biopython's `pairwise2` module:

1. **countMatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment (pairwise2.globalxx) of the same length. It returns the number of positions where the elements of both sequences match.

2. **countMismatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of positions where the elements of the two sequences are different (i.e., they are not gaps, and the characters do not match).

3. **countGapOpens(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap openings in the alignment (a gap is opened when a '-' appears in the sequence).

4. **countGapExtensions(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap extensions (where '-' continues in the alignment after an initial gap is opened).

5. **getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment and returns the alignment score based on the provided scoring scheme: `matchScore` for matches, `mismatchPenalty` for mismatches, `gapOpenPenalty` for opening a gap, and `gapExtensionPenalty` for extending a gap.

```python
# Add your functions here
from Bio import SeqIO
from Bio import pairwise2

def countMatches(s1, s2):
    alignments = pairwise2.align.globalxx(s1, s2)
    aligned_s1, aligned_s2 = alignments[0][0], alignments[0][1]    
    return sum(1 for a, b in zip(aligned_s1, aligned_s2) if a == b)

def countMismatches(s1, s2):
    alignments = pairwise2.align.globalxx(s1, s2)
    aligned_s1, aligned_s2 = alignments[0][0], alignments[0][1]    
    return sum(1 for a, b in zip(aligned_s1, aligned_s2) if a != b and a != '-' and b != '-')

def countGapOpens(s1, s2):
    alignments = pairwise2.align.globalxx(s1, s2)
    aligned_s1, aligned_s2 = alignments[0][0], alignments[0][1]
    gap_opens = 0
    for i in range(1, len(aligned_s1)):
        if aligned_s1[i] == '-' and aligned_s1[i-1] != '-':
            gap_opens += 1
        if aligned_s2[i] == '-' and aligned_s2[i-1] != '-':
            gap_opens += 1
    return gap_opens

def countGapExtensions(s1, s2):
    alignments = pairwise2.align.globalxx(s1, s2)
    aligned_s1, aligned_s2 = alignments[0][0], alignments[0][1]
    gap_extensions = 0
    for i in range(1, len(aligned_s1)):
        if aligned_s1[i] == '-' and aligned_s1[i-1] == '-':
            gap_extensions += 1
        if aligned_s2[i] == '-' and aligned_s2[i-1] == '-':
            gap_extensions += 1
    return gap_extensions

def getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty):
    numMatches = countMatches(s1, s2)
    numMismatches = countMismatches(s1, s2)
    numGapOpens = countGapOpens(s1, s2)
    numGapExtensions = countGapExtensions(s1, s2)
    score = (numMatches * matchScore) + (numMismatches * mismatchPenalty) + (numGapOpens * gapOpenPenalty) + (numGapExtensions * gapExtensionPenalty)    
    return score
```

```python
def longest_increasing_subsequence(arr):
    n = len(arr)
    dp = [1] * n 
    prev = [-1] * n

    for i in range(n):
        for j in range(i):
            if arr[j] < arr[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j

    max_index = dp.index(max(dp))

    lis = []
    while max_index != -1:
        lis.append(arr[max_index])
        max_index = prev[max_index]

    return lis[::-1]


def longest_decreasing_subsequence(arr):
    n = len(arr)
    dp = [1] * n
    prev = [-1] * n

    for i in range(n):
        for j in range(i):
            if arr[j] > arr[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j

    max_index = dp.index(max(dp))

    lds = []
    while max_index != -1:
        lds.append(arr[max_index])
        max_index = prev[max_index]

    return lds[::-1]


def main():
    with open("/Users/makbuk/Downloads/rosalind_lgis.txt", "r") as handle:
        n = int(handle.readline().strip())
        permutation = list(map(int, handle.readline().strip().split()))

    lis = longest_increasing_subsequence(permutation)
    lds = longest_decreasing_subsequence(permutation)

    print(" ".join(map(str, lis)))
    print(" ".join(map(str, lds)))


if __name__ == "__main__":
    main()
```

### lcsq

```python
from Bio import SeqIO

def longest_common_subsequence(s, t):
    m, n = len(s), len(t)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s[i - 1] == t[j - 1]:
            lcs.append(s[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] >= dp[i][j - 1]:
            i -= 1
        else:
            j -= 1
    
    return "".join(reversed(lcs))

lst = []
with open("/Users/makbuk/Downloads/rosalind_lcsq.txt", "r") as handle:
    for seq in SeqIO.parse(handle, "fasta"):
        lst.append(str(seq.seq))

seq1, seq2 = lst[0], lst[1]

print(longest_common_subsequence(seq1, seq2))
```

### edit


```python
def edit_distance(s, t):
    m, n = len(s), len(t)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            substitution_cost = 0 if s[i - 1] == t[j - 1] else 1
            dp[i][j] = min(
                dp[i - 1][j] + 1,
                dp[i][j - 1] + 1,
                dp[i - 1][j - 1] + substitution_cost
            )
    
    return dp[m][n]

sequences = []
with open("/Users/makbuk/Downloads/rosalind_edit.txt", "r") as handle:
    lines = handle.readlines()
    current_seq = []
    for line in lines:
        line = line.strip()
        if line.startswith(">"):
            if current_seq:
                sequences.append("".join(current_seq))
                current_seq = []
        else:
            current_seq.append(line)
    if current_seq:
        sequences.append("".join(current_seq))

seq1, seq2 = sequences[0], sequences[1]

print(edit_distance(seq1, seq2))
```

```python
from Bio import SeqIO

def reconstruct_alignment(s, t):
    m, n = len(s), len(t)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    traceback = [[None] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        dp[i][0] = i
        traceback[i][0] = "UP"
    for j in range(1, n + 1):
        dp[0][j] = j
        traceback[0][j] = "LEFT"
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                cost = 0
            else:
                cost = 1
            dp[i][j] = min(
                dp[i - 1][j] + 1,
                dp[i][j - 1] + 1,
                dp[i - 1][j - 1] + cost
            )
            if dp[i][j] == dp[i - 1][j] + 1:
                traceback[i][j] = "UP"
            elif dp[i][j] == dp[i][j - 1] + 1:
                traceback[i][j] = "LEFT"
            else:
                traceback[i][j] = "DIAG"
    
    aligned_s = []
    aligned_t = []
    i, j = m, n
    while i > 0 or j > 0:
        if traceback[i][j] == "UP":
            aligned_s.append(s[i - 1])
            aligned_t.append("-")
            i -= 1
        elif traceback[i][j] == "LEFT":
            aligned_s.append("-")
            aligned_t.append(t[j - 1])
            j -= 1
        else:
            aligned_s.append(s[i - 1])
            aligned_t.append(t[j - 1])
            i -= 1
            j -= 1

    aligned_s = ''.join(reversed(aligned_s))
    aligned_t = ''.join(reversed(aligned_t))
    
    return dp[m][n], aligned_s, aligned_t

sequences = []
with open("/Users/makbuk/Downloads/rosalind_edta.txt", "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        sequences.append(str(record.seq))

seq1, seq2 = sequences[0], sequences[1]

distance, alignment_s, alignment_t = reconstruct_alignment(seq1, seq2)

print(distance)
print(alignment_s)
print(alignment_t)
```

### ctea

```python
from Bio import SeqIO

MOD = 134_217_727

def count_optimal_alignments(s, t):
    m, n = len(s), len(t)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    path_count = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m + 1):
        dp[i][0] = i
        path_count[i][0] = 1
    for j in range(n + 1):
        dp[0][j] = j
        path_count[0][j] = 1

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            cost = 0 if s[i - 1] == t[j - 1] else 1
            options = [
                (dp[i - 1][j] + 1, path_count[i - 1][j]),
                (dp[i][j - 1] + 1, path_count[i][j - 1]),
                (dp[i - 1][j - 1] + cost, path_count[i - 1][j - 1])
            ]
            min_cost = min(opt[0] for opt in options)
            dp[i][j] = min_cost
            path_count[i][j] = sum(opt[1] for opt in options if opt[0] == min_cost) % MOD

    return path_count[m][n]

sequences = []
with open("/Users/makbuk/Downloads/rosalind_ctea.txt", "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        sequences.append(str(record.seq))

seq1, seq2 = sequences[0], sequences[1]

print(count_optimal_alignments(seq1, seq2))
```
